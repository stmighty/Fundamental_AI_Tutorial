# part3 preprocess (simple)

## 3.1 drop cols 

search command : drop multiple column pandas

link : https://www.geeksforgeeks.org/how-to-drop-one-or-multiple-columns-in-pandas-dataframe/

```python
def drop_cols(df, cols) :
    df_copy = df.copy()
    df_copy = df_copy.drop(cols, axis=1)
    return df_copy
    
```

```python
df_train_without_nan = drop_cols(df_train, all_nan_cols)
df_test_without_nan = drop_cols(df_test, all_nan_cols)
```

```python
df_train_without_nan.head()
```

```python
df_test_without_nan.head()
```

search command : delete all categorical columns pandas

link : https://stackoverflow.com/questions/12725417/drop-non-numeric-columns-from-a-pandas-dataframe

```python
df_train_without_nan_numerical_only = df_train_without_nan.select_dtypes(['number'])
df_train_without_nan_numerical_only.head()
```

```python
df_test_without_nan_numerical_only = df_test_without_nan.select_dtypes(['number'])
df_test_without_nan_numerical_only.head()
```

## 3.2 train test split (train, val)

from your titanic notebook

```python
target_col = "SalePrice"
train_features_X = df_train_without_nan_numerical_only.loc[:, df_train_without_nan_numerical_only.columns != target_col]
train_features_y = df_train_without_nan_numerical_only[target_col]
```

search command : MSE on 2 arrays python

link : https://www.statology.org/mean-squared-error-python/

```python
import numpy as np

def mse(actual, pred): 
    actual, pred = np.array(actual), np.array(pred)
    return np.square(np.subtract(actual,pred)).mean() 

print(mse(y_val, eval_predicted))
```

## 4.2 lightGBM

link : https://www.kaggle.com/code/ashishpatel26/light-gbm-demand-forecasting/notebook

Note : also reference from looloo computer explore folder

```python
import lightgbm as lgb
from lightgbm import LGBMRegressor
from sklearn.model_selection import TimeSeriesSplit
from sklearn.metrics import mean_squared_log_error

def model(train_x, train_y, test_x, test_y, test_df):

    params = {
        'nthread': 10,
        'max_depth': 5,  # Reduce max_depth for simplicity
        'boosting_type': 'gbdt',
        'objective': 'regression_l1',
        'metric': 'mape',
        'num_leaves': 8,  # Ensure num_leaves is suitable for max_depth
        'learning_rate': 0.1,
        'feature_fraction': 0.8,
        'bagging_fraction': 0.8,
        'bagging_freq': 5,
        'lambda_l1': 0.0,  # Simplify regularization
        'lambda_l2': 0.0,
        'min_child_weight': 1,
        'min_split_gain': 0.0
    }
    
    lgb_train = lgb.Dataset(train_x, train_y)
    lgb_valid = lgb.Dataset(test_x, test_y, reference=lgb_train)

    model = lgb.train(
        params, 
        lgb_train, 
        num_boost_round=1000, 
        valid_sets=[lgb_train, lgb_valid], 
        callbacks=[lgb.early_stopping(stopping_rounds=50), lgb.log_evaluation(period=50)]
    )
    
    y_test = model.predict(test_df)
    return y_test, model


SalePrice_light_gbm, model = model(X_train, y_train, X_Val, y_val, df_test_without_nan_numerical_only)
```
